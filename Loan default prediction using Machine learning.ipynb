# Feature Selection Using Lasso Regression 

```python
#importing libraries 
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso

y=Y 

# Calling Lasso
pipeline = Pipeline([('scaler',StandardScaler()),('model',Lasso())])
search = GridSearchCV(pipeline,{'model__alpha':np.arange(0.01,0.1,0.01)}, cv = 5, scoring="neg_mean_squared_error",verbose=3 )
search.fit(X,y)

# Choosen Alpha 
print('\n The Best Parameter:',search.best_params_)
features = X_train.columns

#Cofficients 
coefficients = search.best_estimator_.named_steps['model'].coef_
print(coefficients )
importance = np.abs(coefficients)
print(importance )


print('\n The Features Selected:',np.array(features)[importance > 0])


print('\n The Features that can be ignored:',np.array(features)[importance == 0])


```

# Performance metrics

```python
ypred=classifier.predict(X_test)
```

```python
#accuracy
#Confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, ypred)
print(cm)

from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

print('Precision score:', precision_score(y_test, ypred))
print('Recall score:', recall_score(y_test, ypred))
print('Accuracy score:', accuracy_score(y_test, ypred))
print('F1 score:', f1_score(y_test, ypred))
```

```python
# Precision-Recall Curve

from sklearn.metrics import precision_recall_curve

# S1. Get probability

logreg_prob = classifier.predict_proba(X_test)[: , 1]

# S2. Collect Necessary Measures

precision, recall, threshold = precision_recall_curve(y_test, logreg_prob)

# S3. Plot Figure

plt.plot(recall, precision)
# plt.plot(recall_1, precision_1)
plt.xlabel("recall")
plt.ylabel("precision")
plt.title("Precision-Recall Curve")
```

```python
from sklearn.metrics import roc_auc_score

# we have obtained the predicted probability in the previous step: y_pred_proba
print("AUC for logistic regression:",   roc_auc_score(y_test, logreg_prob))
```

```python
#ROC Curve

from sklearn.metrics import roc_curve

# Predicted Probability: y_pred_proba
# Get Measures
fpr, tpr, threshold = roc_curve(y_test, logreg_prob)

# Generate Figure

plt.plot(fpr, tpr)
plt.xlabel("fpr")
plt.ylabel("tpr")
plt.title("ROC")
```
